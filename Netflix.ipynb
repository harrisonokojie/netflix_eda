**Drop Null Values**

```python
netflix_copy = netflix.copy()

netflix_copy = netflix_copy.dropna()

netflix_copy.shape
```

**Duplicate Values**

```python
print(f"Duplicate values found : {netflix_copy.duplicated().sum()}")
```

**Rating Types**

```python
netflix_copy['rating'].value_counts()
```

**Converting Date Time Format**

```python
netflix_copy['date_added'] = pd.to_datetime(netflix_copy['date_added'])
netflix_copy['day_added'] = netflix_copy['date_added'].dt.day
netflix_copy['year_added'] = netflix_copy['date_added'].dt.year
netflix_copy['month_added']= netflix_copy['date_added'].dt.month
netflix_copy['year_added'].astype(int);
netflix_copy['day_added'].astype(int);
netflix_copy.head(10)
```

**Visualization of Data**

Movies against Tv Shows

```python
sns.countplot(netflix_copy, x ='type')

plt.xticks(rotation = 45)
plt.xlabel('Content Type')
plt.ylabel('Count')
plt.title('Distribution of Movies and Tv Shows on Netflix')
plt.show()
```
